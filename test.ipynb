```python
import pickle
d = {}
files = os.listdir()
```

```python
for file in files:
    try:
        pickle_in = open(file,"rb")
        res = pickle.load(pickle_in)
        d[str(res[-1][0]) + ", " + str(res[-1][1])] = len(res) - 6
    except:
        os.remove(file)
        print(file)
    
```

```python
len(d)
```

```python
norm_dims = (275, 373)
rightmost,leftmost,topmost,bottommost = (-121.21243727715039, -122.20127766290896, 37.48908544352257, 36.888272429602885)
delta_lat = (topmost - bottommost) / (norm_dims[0] - 1)
delta_long = (rightmost - leftmost) / (norm_dims[1] - 1)
```

```python
added = 0
for i in range(norm_dims[0]):
    latitude = topmost - i * delta_lat
    for j in range(norm_dims[1]):
        longitude = rightmost - j * delta_long
        key = str(latitude) + ", " + str(longitude)
        if key not in d:
            d[key] = -1
            added += 1
            
print(added)
            
            
        
```

```python
population = []
latitude = []
longitude = []

for key in keys:
    population.append(d[key])
    latitude.append(key.split(", ")[0])
    longitude.append(key.split(", ")[1])
```

```python
import pandas as pd

# initialise data of lists.
data = {'Population': population,
		'Latitude':latitude,
        'Longitude':longitude}

# Creates pandas DataFrame.
df = pd.DataFrame(data)

# print the data
df
```
